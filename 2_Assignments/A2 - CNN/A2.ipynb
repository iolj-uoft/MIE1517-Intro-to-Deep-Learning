### Import Libraries

````python
# import libraries
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision
from torch.utils.data.sampler import SubsetRandomSampler
import os
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, random_split
import matplotlib.pyplot as plt
````

### Split Dataset and Fit with the Format for `ImageFolder`

````python
import shutil
import random

random.seed(1)
dataset_path = "Lab_3b_Gesture_Dataset Revised"
output_path = "Processed_Lab_3b_Gesture_Dataset Revised"

train_ratio = 0.6 # 60% for training
valid_ratio = 0.2 # 20% for validation
test_ratio = 0.2 # 20% for testing

# Remove existing directories if they exist
if os.path.exists(output_path):
    shutil.rmtree(output_path)

# Create three subfolders to contain each dataset
for split in ["train", "valid", "test"]:
    os.makedirs(os.path.join(output_path, split), exist_ok=True)

for class_name in os.listdir(dataset_path):
    # iterate through the name of the subfolders (class) and set the path to that class
    class_path = os.path.join(dataset_path, class_name) 
    
    if os.path.isdir(class_path):
        images = [f for f in os.listdir(class_path) if f.endswith((".jpg"))] # iterate all files in the class folder and store the images into a list
        random.shuffle(images) # randomize the order of the images
        
        # define the len for each dataset for indexing
        len_train = int(len(images) * train_ratio)
        len_valid = int(len(images) * valid_ratio)
        len_test = int(len(images) * test_ratio)
        
        # split the dataset
        train_images = images[:len_train]
        valid_images = images[len_train:(len_train + len_valid)]
        test_images = images[(len_train + len_valid):]
        
        # create subfolders for processed data and copy them to the directory
        for split_name, split_images in zip(["train", "valid", "test"], [train_images, valid_images, test_images]):
            class_output_path = os.path.join(output_path, split_name, class_name)
            os.makedirs(class_output_path, exist_ok=True)
            
            for image in split_images:
                shutil.copy(os.path.join(class_path, image), os.path.join(class_output_path, image))
        
````

````python
# define training and test data directories
data_dir = 'Processed_Lab_3b_Gesture_Dataset Revised'
train_dir = os.path.join(data_dir, 'train/')
val_dir = os.path.join(data_dir, 'valid/')
test_dir = os.path.join(data_dir, 'test/')

# classes are folders in each directory with these names
classes = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']
print(classes)
````

````python

# load and transform data using ImageFolder

# resize all images to 224 x 224
data_transform = transforms.Compose([transforms.RandomResizedCrop(224),
                                      transforms.ToTensor()])

train_data = datasets.ImageFolder(train_dir, transform=data_transform)
val_data = datasets.ImageFolder(val_dir, transform=data_transform)
test_data = datasets.ImageFolder(test_dir, transform=data_transform)

# print out some data stats
print('Num training images: ', len(train_data))
print('Num validation images: ', len(val_data))
print('Num testing images: ', len(test_data))
````

````python
# define dataloader parameters
batch_size  = 20
num_workers = 4

# prepare data loaders
train_loader = torch.utils.data.DataLoader(train_data, batch_size=batch_size,
                                           num_workers=num_workers, shuffle=True)
val_loader = torch.utils.data.DataLoader(val_data, batch_size=batch_size,
                                          num_workers=num_workers, shuffle=True)
````

````python
# Visualize some sample data

# obtain one batch of training images
dataiter = iter(train_loader)
images, labels = next(dataiter)
images = images.numpy() # convert images to numpy for display

# plot the images in the batch, along with the corresponding labels
fig = plt.figure(figsize=(25, 4))
for idx in np.arange(20):
    ax = fig.add_subplot(2, 20//2, idx+1, xticks=[], yticks=[])
    plt.imshow(np.transpose(images[idx], (1, 2, 0)))
    ax.set_title(classes[labels[idx]])
````
