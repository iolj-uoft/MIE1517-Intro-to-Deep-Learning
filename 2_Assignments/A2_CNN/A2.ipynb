### Split Dataset and Fit with the Format for `ImageFolder`

````python
random.seed(100)
dataset_path = "Lab_3b_Gesture_Dataset Revised"
output_path = "Processed_Lab_3b_Gesture_Dataset Revised"

train_ratio = 0.6  # 60% for training
valid_ratio = 0.2  # 20% for validation

test_ratio = 0.2  # 20% for testing

# Remove existing directories if they exist
if os.path.exists(output_path):
    shutil.rmtree(output_path)

# Create three subfolders to contain each dataset
for split in ["train", "valid", "test"]:
    os.makedirs(os.path.join(output_path, split), exist_ok=True)

# Group images by person
class_person_images = defaultdict(lambda: defaultdict(list))

for class_name in os.listdir(dataset_path):
    class_path = os.path.join(dataset_path, class_name)
    if os.path.isdir(class_path):
        for image in os.listdir(class_path):
            if image.endswith(".jpg"):
                person_id = image.split("_")[0]  # Extract person ID from filename prefix
                class_person_images[class_name][person_id].append(image)

# Split by person instead of individual images
for class_name, person_images in class_person_images.items():
    person_ids = list(person_images.keys())
    random.shuffle(person_ids)
    
    len_train = int(len(person_ids) * train_ratio)
    len_valid = int(len(person_ids) * valid_ratio)
    
    train_ids = person_ids[:len_train]
    valid_ids = person_ids[len_train:len_train + len_valid]
    test_ids = person_ids[len_train + len_valid:]  # Remaining persons go to test set
    
    splits = {"train": train_ids, "valid": valid_ids, "test": test_ids}
    
    for split_name, split_ids in splits.items():
        class_output_path = os.path.join(output_path, split_name, class_name)
        os.makedirs(class_output_path, exist_ok=True)
        
        for person_id in split_ids:
            for image in person_images[person_id]:
                shutil.copy(os.path.join(dataset_path, class_name, image), os.path.join(class_output_path, image))
````

### Part (i) Convolutional Network [2pt MODEL]

Build a convolutional neural network model that takes the (224x224 RGB) image as input, and predicts the gesture
letter. Your model should be a subclass of nn.Module. Explain your choice of neural network architecture: how

````python
# TO BE COMPLETED
class SignLanguageClassifier(nn.Module):
    def __init__(self):
        super(SignLanguageClassifier, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, 3, padding=1) # in_channels, out_chanels, kernel_size, N x 32 x 224 x 224
        self.conv2 = nn.Conv2d(32, 64, 3, padding=1) # N x 64 x 224 x 224
        self.pool1 = nn.MaxPool2d(2, 2) # N x 64 x 112 x 112
        self.conv3 = nn.Conv2d(64, 128, 3, padding=1) # N x 128 x 112 x 112
        self.conv4 = nn.Conv2d(128, 128, 3, padding=1) # N x 128 x 112 x 112
        self.pool2 = nn.MaxPool2d(2, 2) # N x 128 x 56 x 56
        self.conv5 = nn.Conv2d(128, 256, 3, padding=1) # N x 256 x 56 x 56
        self.conv6 = nn.Conv2d(256, 256, 3, padding=1) # N x 256 x 56 x 56
        self.pool3 = nn.MaxPool2d(2, 2) # N x 256 x 28 x 28
        self.fc1 = nn.Linear(256 * 28**2, 512) 
        self.fc2 = nn.Linear(512, 256)
        self.fc3 = nn.Linear(256, 9)
        
    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = self.pool1(x)
        x = F.relu(self.conv3(x))
        x = F.relu(self.conv4(x))
        x = self.pool2(x)
        x = F.relu(self.conv5(x))
        x = F.relu(self.conv6(x))
        x = self.pool3(x)
        x = x.view(-1, 256 * 28**2)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
````

````python
class ImprovedSignLanguageClassifier(nn.Module):
    def __init__(self):
        super(ImprovedSignLanguageClassifier, self).__init__()
        self.conv1 = nn.Conv2d(3, 64, 3, padding=1)
        self.bn1 = nn.BatchNorm2d(64)
        
        self.conv2 = nn.Conv2d(64, 128, 3, padding=1)
        self.bn2 = nn.BatchNorm2d(128)
        
        self.conv3 = nn.Conv2d(128, 256, 3, padding=1)
        self.bn3 = nn.BatchNorm2d(256)
        
        self.conv4 = nn.Conv2d(256, 512, 3, padding=1)
        self.bn4 = nn.BatchNorm2d(512)

        self.pool = nn.MaxPool2d(2, 2)
        self.global_pool = nn.AdaptiveAvgPool2d((1, 1))

        self.fc1 = nn.Linear(512, 128)
        self.fc2 = nn.Linear(128, 9)

    def forward(self, x):
        x = self.pool(F.relu(self.bn1(self.conv1(x))))
        x = self.pool(F.relu(self.bn2(self.conv2(x))))
        x = self.pool(F.relu(self.bn3(self.conv3(x))))
        x = self.pool(F.relu(self.bn4(self.conv4(x))))
        x = self.global_pool(x)
        x = x.view(x.shape[0], -1)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
````

### Part (ii) Training Code [1pt MODEL]

Write code that trains your neural network given some training data. Your training code should make it easy

````python
def get_accuracy(model, train=False):
    # Use the correct DataLoader
    loader = train_loader if train else val_loader

    correct = 0
    total = 0

    with torch.no_grad():
        for imgs, labels in loader:  # Use train_loader or val_loader
            # To Enable GPU Usage
            if use_cuda and torch.cuda.is_available():
                imgs = imgs.cuda()
                labels = labels.cuda()
                
            output = model(imgs)
            pred = output.argmax(1)  # Get class predictions

            correct += (pred == labels).sum().item()
            total += imgs.shape[0]
            
            # Print first 5 predictions for debugging
            print("Predictions:", pred[:10].cpu().numpy())
            print("True Labels:", labels[:10].cpu().numpy())

    return correct / total
````

````python
# TO BE COMPLETED
def train(model, data, batch_size=64, num_epochs=1, lr=0.005):
    train_loader = torch.utils.data.DataLoader(data, batch_size=batch_size, shuffle=True, num_workers=0, pin_memory=True)
    loss_function = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)

    iters, losses, train_acc, val_acc = [], [], [], []

    # training
    n = 0 # the number of iterations
    for epoch in range(num_epochs):
        for batch_idx, (imgs, labels) in enumerate(train_loader):
            #To Enable GPU Usage
            if use_cuda and torch.cuda.is_available():
              imgs = imgs.cuda()
              labels = labels.cuda()
            print(f"Batch {batch_idx}: {labels.cpu().numpy()}")
            if batch_idx == 5: break  # Stop after 5 batches
            out = model(imgs)             # forward pass
            loss = loss_function(out, labels) # compute the total loss
            loss.backward()               # backward pass (compute parameter updates)
            # for name, param in model.named_parameters():
            #   if param.grad is not None:
            #       print(f"{name}: Mean Grad = {param.grad.abs().mean().item()}")  
            optimizer.step()              # make the updates for each parameter
            optimizer.zero_grad()         # a clean up step for PyTorch
            if batch_idx % 10 == 0:  # Print loss every 10 batches
              print(f"Batch {batch_idx}, Loss: {loss.item()}")
            # save the current training information
            iters.append(n)
            losses.append(float(loss)/batch_size)             # compute *average* loss
            train_acc.append(get_accuracy(model, train=True)) # compute training accuracy
            val_acc.append(get_accuracy(model, train=False))  # compute validation accuracy
            n += 1

    # plotting
    plt.title("Training Curve")
    plt.plot(iters, losses, label="Train")
    plt.xlabel("Iterations")
    plt.ylabel("Loss")
    plt.show()

    plt.title("Training Curve")
    plt.plot(iters, train_acc, label="Train")
    plt.plot(iters, val_acc, label="Validation")
    plt.xlabel("Iterations")
    plt.ylabel("Training Accuracy")
    plt.legend(loc='best')
    plt.show()

    print("Final Training Accuracy: {}".format(train_acc[-1]))
    print("Final Validation Accuracy: {}".format(val_acc[-1]))
````

For many image classification tasks, it is generally not a good idea to train a very large deep neural network
model from scratch due to the enormous compute requirements and lack of sufficient amounts of training
data.

One of the better options is to try using an existing model that performs a similar task to the one you need
to solve. This method of utilizing a pre-trained network for other similar tasks is broadly termed **Transfer
Learning**. In this assignment, we will use Transfer Learning to extract features from the hand gesture
images. Then, train a smaller network to use these features as input and classify the hand gestures.

As you have learned from the CNN lecture, convolution layers extract various features from the images which
get utilized by the fully connected layers for correct classification. AlexNet architecture played a pivotal
role in establishing Deep Neural Nets as a go-to tool for image classification problems and we will use an
ImageNet pre-trained AlexNet model to extract features in this assignment.
